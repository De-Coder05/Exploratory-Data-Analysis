```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df=pd.read_csv("/content/nbaDataset.csv")
df1=pd.DataFrame(df)
df1
```

```python
df.count()
df=df.dropna()
df=df.drop(['Number'],axis=1)
df
```

```python
df=df.drop_duplicates()
```

```python
df.Team.value_counts().nlargest(40).plot(kind="bar",figsize=(10,5))
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Age'])
```

```python
plt.figure(figsize=(10,6))
d=sns.scatterplot(x=df['Age'],y=df['Salary'])
plt.title("Scatterplot")
plt.ylabel("Salary in Crore")
```

```python
c=df.corr(numeric_only=True)
sns.heatmap(c,annot=True)
```
